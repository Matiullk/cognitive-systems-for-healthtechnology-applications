```python
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
import pandas as pd
```

```python
x = np.arange(1001)
print(x)
```

```python
y = np.log10(x+1)
print(y)
```

```python
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of y = log(x+1)')
plt.grid()
plt.show()
```
